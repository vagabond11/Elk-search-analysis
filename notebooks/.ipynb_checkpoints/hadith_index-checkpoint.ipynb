```python
from elasticsearch import helpers, Elasticsearch
import json
import codecs

es = Elasticsearch([{'host': 'localhost', 'port': '9200'}],timeout=30, max_retries=10, retry_on_timeout=True)
es.ping()
```

```python
hadith = codecs.open('data/json_hadith.json', 'r', 'utf-8')

json_hadith = json.load(hadith)
json_hadith[2]
```

```python
hadit = []

for obj in json_hadith:
      hadit.append({
        "source_arabic": obj['source'],
        "source_english": obj['source_english'],
        "chapter_no": obj['chapter_no'],
        "chapter": obj['chapter'],
        "hadith_arabic": obj['text_ar'],
        "hadith_english": obj['text_en'],

    })

print(hadit[2])
```

```python
mapping ={
    "settings" : {
       "number_of_shards":1,
       "number_of_replicas":1,

    "analysis":{
   "filter":{
      "arabic_stop":{
         "type":"stop",
         "stopwords_path":"D:/Elastic/elasticsearch/config/merged-stopwords.txt" # change to D:/Elastic/elasticsearch/config/merged-stopwords.txt
      },
      "arabic_keywords":{
         "type":"keyword_marker",
         "keywords_path":"D:/Elastic/elasticsearch/config/keywordmarker-words.txt" # change to D:/Elastic/elasticsearch/config/keywordmarker-words.txt
      },
      "stop_complex_queries":{
         "type":"stop",
         "stopwords_path":"D:\Elastic\elasticsearch\config\merged-stopwords-1.txt" #copy the txt file to the corresponding ES directory
      },
      "arabic_stemmer":{
         "type":"stemmer",
         "language":"arabic"
      },
      "arabic_shingle":{
         "type":"shingle",
         "min_shingle_size":2,
         "max_shingle_size":3
      }
   },
   "analyzer":{
      "rebuilt_arabic":{
         "tokenizer":"standard",
         "filter":[
            "arabic_stop",
            "arabic_normalization",
            "arabic_keywords",
            "arabic_shingle",
            "arabic_stemmer"
         ]
      },
       "stem_noshingle":{
         "tokenizer":"standard",
         "filter":[
            "arabic_stop",
            "arabic_normalization",
            "arabic_keywords",
            "arabic_stemmer" #for one word aggregations 
         ]
      },
      "shingle_nostem":{
         "tokenizer":"standard",
         "filter":[
            "arabic_stop",
            "arabic_normalization",
            "arabic_shingle"
         ]
      },
      "simple": {
          "tokenizer":"standard",
         "filter":[
            "arabic_stop",
            "arabic_normalization"
         ]
      },
      "comp_query": {
          "tokenizer":"standard",
         "filter":[
            "stop_complex_queries",
            "arabic_normalization"
         ]
      }
   }
}
    },
"mappings":{
   "properties":{
      "source_arabic":{
         "type":"text",
          "analyzer" : "simple",
          "fielddata" : "true",
          "fields":{
            "keyword":{
               "type":"keyword"
            }
         }
      },
       "source_english":{
         "type":"text",
          "fields":{
            "keyword":{
               "type":"keyword"
            }
         }
      },
      "chapter_no":{
         "type":"integer"
      },
       "chapter":{
         "type":"text",
         "analyzer":"simple" 
      },
      "hadith_arabic":{
         "type":"text",
         "analyzer":"rebuilt_arabic",
         "fielddata":"true",
         "fields" : {
            "simple":{
               "type":"text",
               "analyzer":"simple",
               "fielddata":"true"
            },
             "stem_noshingle":{
               "type":"text",
               "analyzer":"stem_noshingle",
               "fielddata":"true"
             },
             "shingle_nostem":{
               "type":"text",
               "analyzer":"shingle_nostem",
               "fielddata":"true"
             },
             "comp_query":{
               "type":"text",
               "analyzer":"comp_query",
               "fielddata":"true"
             },
            "keyword":{
               "type":"keyword"
            }
         }
      },
      "hadith_english":{
          "type":"text",
          "analyzer":"english"
          }
   }
}
}
```

```python
es.indices.create(index="hadith_no_mapping",ignore=400)
```

```python
helpers.bulk(es,index='hadith_no_mapping', actions=hadit)
```

```python
es.indices.create(index="hadith_with_mapping", body = mapping ,ignore=400)
```

```python
helpers.bulk(es,index='hadith_with_mapping', actions=hadit, ignore=400)
```

```python
#Verify creation of the Hadith indices
res= es.indices.get_alias("*")
for nomindex in res:
    print(nomindex)
```
